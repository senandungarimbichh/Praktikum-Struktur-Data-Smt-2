```python
# Membuat kelas node dan linked list
class Node:
    def init (self, data) :
        self.data = data
        self.next = None
```

```python
# Implementasi singly linked list
class SinglyLinkedList:
    def __init__(self):
      self.head = None
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
    def display(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

```python
# Implementasi Doubly Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
    def display(self) :
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
```

```python
# Implementasi circular Linked List
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return
        # Corrected indentation here
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    def display(self):
        if not self.head:
            print("List is empty")
            return
        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print('(Back to head)')
```

```python
# LATIHAN 1 #
#implementasi singly linked list
#Definisi Node dalam Linked list
class TaskNode:
  def __init__(self,task_name):#menyimpan mata kuliah
    self.task_name = task_name
    self.next = None #pointer ke mata kuliah berikutnya
#Definisi Linked list untuk mata kuliah hari jumat
class TaskList:
  def __init__(self):
    self.head = None #awalnya tidak ada mata kuliah (daftar kosong)
  #menambahkan mata kuliah ke linked list
  def add_task(self, task_name):
    new_task = TaskNode(task_name)
    if not self.head:
      self.head = new_task #jika list kosong, mata kuliah pertama menjadi head
    else:
      current = self.head
      while current.next: #mencari mode terakhir dalam list
        current = current.next
      current.next = new_task #menambahkan mata kuliah di akhir list
  #menampilkan daftar mata kuliah
  def show_task(self):
      if not self.head:
        print("\n Tidak ada mata kuliah hari ini!")
        return
      print("\n Daftar mata kuliah hari ini:")
      current = self.head
      count = 1
      while current:
        print(f"{count}. {current.task_name}")
        current = current.next
        count += 1

  #============== program utama (main)===============#

  #membuat objek daftar mata kuliah
task_list = TaskList()
  #meminta pengguna memasukkan jumlah mata kuliah
num_task = int(input("masukkan jumlah mata kuliah sesuai jumlah npm : "))
  #loop untuk memasukkan mata kuliah ke dalam linked list
for i in range(1, num_task + 1):
    task_name = input(f"masukkan mata kuliah ke-{i}: ")
    task_list.add_task(task_name)
 #menampilkan hasil daftar mata kuliah
task_list.show_task()
```

```python
# LATIHAN 2 #
# Definisi Node dalam Doubly Linked List
class PageNode:
  def __init__(self, url): # Corrected: def_init_ to def __init__
    self.url = url # URL halaman web
    self.next = None #Pointer ke halaman berikutnya
    self.prev = None # Pointer ke halamann sebelumnya
# Definisi Doubly Linked List untuk Riwayat Peramban
class BrowserHistory:
  def __init__(self): # Corrected: def_init_ to def __init__
      self.current = None # Pointer ke halaman yang sedang dibuka
      # Menambahkan halaman baru ke dalam riwayat
  def visit_page(self, url):
      new_page = PageNode(url)
      if not self.current:
          self.current = new_page # Jika riwayat kosong, halaman ini menjadi halaman pertama
      else:
          new_page.prev = self.current
          self.current.next = new_page
          self.current = new_page # Pindahkan posisi ke halaman baru
      # Menampilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
        print("\n Riwayat navigasi folder kosong.")
        return
    print("\n Riwayat navigasi folder:")
    temp = self.current
    while temp.prev: # Kembali ke halaman pertama
        temp = temp.prev
    count = 1 # Corrected: Indentation aligned with the while loop
    while temp:
        print(f"{count}. {temp.url}")
        temp = temp.next
        count += 1
      # Navigasi mundur ke halaman sebelumnya
  def go_back(self):
      if self.current and self.current.prev:
          self.current = self.current.prev
          print(f"\n Anda kembali ke: {self.current.url}")
      else:
          print("\n Tidak bisa kembali, ini adalah halaman pertama!")
      # Navigasi maju ke  halaman berikutnya
  def go_forward(self): # Corrected: Function name to go_forward
    if self.current and self.current.next: # Corrected: self.current.prev to self.current.next
      self.current = self.current.next # Corrected: self.current.prev to self.current.next
      print(f"\n Anda maju ke : {self.current.url}")
    else:
      print("\n Tidak bisa maju, ini adalah halaman terakhir!")
# ====== Program Utama (Main) ====== #
#  Membuat objek riwayat peramban
history = BrowserHistory()
# Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("\nMasukkan jumlah halaman yang dikunjungi: "))
# Loop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1): # Corrected: semicolon to colon
    url = input(f"Masukkan URL halaman ke-{i}: ") # Corrected: f: to f"
    history.visit_page(url)
# Menampilkan riwayat yang telah dimasukkan
history.show_history()
# Menu navigasi pengguna
while True:
    command = input("\n Tekan 'b' untuk kembali, Tekan 'f' untuk maju, Tekan 'x' untuk keluar.\n Masukkan perintah: ").lower()
    if command == 'b':
        history.go_back()
    elif command == 'f':
        history.go_forward() # Corrected: go_back to go_forward
    elif command == 'x':
        print("\n Keluar dari riwayat navigasi folder.")
        break
    else:
      print("\n perintah tidak dikenal.")
```

```python
# LATIHAN 3 #
# Definisi Node dalam Singly Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya

# Definisi Singly Linked List dengan Berbagai Operasi
class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang list!")
                return
            temp = temp.next
        if not temp:
            print("Posisi melebihi panjang list!")
            return
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next
        return

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None
    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang list!")
                return
            temp = temp.next
        if not temp:
            print("Posisi melebihi panjang list!")
            return
        new_node.next = temp.next
        temp.next = new_node
    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print("⚠️Elemen {key} tidak ditemukan dalam daftar!")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print("Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end="->")
            temp = temp.next
        print("None")

# ======== Program Utama (Main) ========

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))

for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\n Pilih Operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```
