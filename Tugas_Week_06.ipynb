```python
def quick_sort(arr, reverse=False):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    if reverse:
        return quick_sort(right, reverse) + middle + quick_sort(left, reverse)
    else:
        return quick_sort(left, reverse) + middle + quick_sort(right, reverse)
# Contoh data produk
produk = [
    {"nama": "Kemeja Flanel", "harga": 200000, "rating": 4.5},
    {"nama": "Kaos Polos", "harga": 75000, "rating": 4.0},
    {"nama": "Jaket Denim", "harga": 350000, "rating": 4.8},
    {"nama": "Celana Chino", "harga": 180000, "rating": 4.2},
    {"nama": "Sweater Rajut", "harga": 250000, "rating": 4.7},
]

# Pilihan pengguna
print("Pilih metode pengurutan:")
print("1. Berdasarkan harga")
print("2. Berdasarkan rating")
pilihan = int(input("Masukkan pilihan (1/2): "))

if pilihan == 1:
    harga_produk = [p["harga"] for p in produk]
    sorted_harga_asc = quick_sort(harga_produk.copy())
    sorted_harga_desc = quick_sort(harga_produk.copy(), reverse=True)

    print("\nHarga setelah Quick Sort (termurah ke termahal):")
    for harga in sorted_harga_asc:
        for p in produk:
            if p["harga"] == harga:
                print(f"{p['nama']}: Rp{p['harga']}")
                break

    print("\nHarga setelah Quick Sort (termahal ke termurah):")
    for harga in sorted_harga_desc:
        for p in produk:
            if p["harga"] == harga:
                print(f"{p['nama']}: Rp{p['harga']}")
                break

elif pilihan == 2:
    rating_produk = [p["rating"] for p in produk]
    sorted_rating_asc = quick_sort(rating_produk.copy())
    sorted_rating_desc = quick_sort(rating_produk.copy(), reverse=True)

    print("\nRating setelah Quick Sort (terendah ke tertinggi):")
    for rating in sorted_rating_asc:
        for p in produk:
            if p["rating"] == rating:
                print(f"{p['nama']}: Rating {p['rating']}")
                break

    print("\nRating setelah Quick Sort (tertinggi ke terendah):")
    for rating in sorted_rating_desc:
        for p in produk:
            if p["rating"] == rating:
                print(f"{p['nama']}: Rating {p['rating']}")
                break

else:
    print("Pilihan tidak valid. Silakan jalankan ulang program.")
```
