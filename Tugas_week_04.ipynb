```python
class AntrianTiket:
    def __init__(self, capacity=10):
        self.queue = []
        self.capacity = capacity

    def enqueue(self, ticket_id):
        if '7' in ticket_id:
            print(f"🎟️ VIP Customer {ticket_id}, Anda telah mendapatkan kursi tanpa harus antrian!")
        elif len(self.queue) >= self.capacity:
            print(f"⚠️ Antrian penuh! {ticket_id} harap menunggu sebentar.")
        else:
            self.queue.append(ticket_id)
            print(f"✅ Pelanggan {ticket_id} masuk ke antrian.")

    def dequeue(self):
        if self.queue:
            pelanggan = self.queue.pop(0)
            print(f"🎬 Pelanggan {pelanggan} telah membeli tiket.")
        else:
            print("⚠️ Antrian kosong, tidak ada pelanggan yang bisa dilayani.")

    def front(self):
        print(f"👤 Pelanggan pertama dalam antrian: {self.queue[0]}" if self.queue else "⚠️ Antrian kosong, tidak ada pelanggan.")

    def rear(self):
        print(f"🎟️ Pelanggan terakhir dalam antrian: {self.queue[-1]}" if self.queue else "⚠️ Antrian kosong, tidak ada pelanggan.")

    def show_queue(self):
        if self.queue:
            print("\n📌 Antrian saat ini: " + " -> ".join(self.queue))
        else:
            print("⚠️ Antrian kosong.")

# Inisialisasi antrian
antrian = AntrianTiket()

while True:
    print("\n<------------------------->")
    print("🎟️ MENU TIKET:")
    print("1️⃣ Pesan Tiket")
    print("2️⃣ Layani Pelanggan")
    print("3️⃣ Lihat Antrian")
    print("4️⃣ Lihat Antrian Pertama")
    print("5️⃣ Lihat Antrian Terakhir")
    print("6️⃣ Keluar")
    print("<------------------------->")

    pilihan = input("📌 Pilih Menu Tiket: ")

    if pilihan == "1":
        ticket_id = input("🔢 Masukkan kode pelanggan: ").upper()
        antrian.enqueue(ticket_id)

    elif pilihan == "2":
        antrian.dequeue()

    elif pilihan == "3":
        antrian.show_queue()

    elif pilihan == "4":
        antrian.front()

    elif pilihan == "5":
        antrian.rear()

    elif pilihan == "6":
        print("🙏 Terima kasih! Program selesai.")
        break

    else:
        print("⚠️ Pilihan tidak valid, silakan coba lagi.")
```

```python
class AntrianFotokopi:
    def __init__(self):
        self.queue = []

    def is_empty(self):
        return len(self.queue) == 0

    def enqueue(self, order_id, num_pages):
        num_pages = int(num_pages)

        if num_pages < 10:
            print(f"✅ Pesanan {order_id} ({num_pages} halaman) selesai, pesanan bisa langsung diambil.")
            return

        if num_pages > 100:
            print(f"⚠️ Pesanan {order_id} ({num_pages} halaman) akan diproses dalam dua tahap.")
            self.queue.append((order_id, 100))
            self.queue.append((order_id, num_pages - 100))
        else:
            self.queue.append((order_id, num_pages))

        print(f"📄 Pesanan {order_id} ({num_pages} halaman) masuk ke antrian.")

    def dequeue(self):
        if self.is_empty():
            print("⚠️ Antrian kosong, tidak ada pelanggan yang bisa dilayani.")
        else:
            pelanggan = self.queue.pop(0)
            print(f"🎉 Pesanan {pelanggan[0]} ({pelanggan[1]} halaman) telah selesai diproses.")

    def front(self):
        if self.is_empty():
            print("⚠️ Antrian kosong.")
        else:
            print(f"👤 Pelanggan pertama dalam antrian: {self.queue[0][0]} ({self.queue[0][1]} halaman)")

    def rear(self):
        if self.is_empty():
            print("⚠️ Antrian kosong.")
        else:
            print(f"📜 Pelanggan terakhir dalam antrian: {self.queue[-1][0]} ({self.queue[-1][1]} halaman)")

    def show_queue(self):
        if self.is_empty():
            print("⚠️ Antrian kosong.")
        else:
            print("\n📌 Antrian saat ini: " + " -> ".join(f"{order[0]} ({order[1]} halaman)" for order in self.queue))

# Inisialisasi antrian
antrian = AntrianFotokopi()

while True:
    print("\n<------------------------->")
    print("📑 MENU PELAYANAN FOTOKOPI")
    print("1️⃣ Pesan Fotokopi")
    print("2️⃣ Layani Pelanggan")
    print("3️⃣ Lihat Antrian")
    print("4️⃣ Lihat Pelanggan Pertama")
    print("5️⃣ Lihat Pelanggan Terakhir")
    print("6️⃣ Keluar")
    print("<------------------------->")

    pilihan = input("📌 Pilih Menu Pelayanan: ")

    if pilihan == "1":
        order_id = input("🔢 Masukkan ID Pemesanan: ").upper()
        num_pages = input("📄 Masukkan jumlah halaman: ")

        if not num_pages.isdigit() or int(num_pages) <= 0:
            print("⚠️ Jumlah halaman harus berupa angka positif!")
            continue

        antrian.enqueue(order_id, num_pages)

    elif pilihan == "2":
        antrian.dequeue()

    elif pilihan == "3":
        antrian.show_queue()

    elif pilihan == "4":
        antrian.front()

    elif pilihan == "5":
        antrian.rear()

    elif pilihan == "6":
        print("🙏 Terima kasih! Sampai jumpa kembali.")
        break

    else:
        print("⚠️ Pilihan tidak valid, silakan coba lagi.")
```

```python
from collections import deque

class Antrian_Layanan_Parkir:
    def __init__(self, capacity=15):
        self.queue = deque()
        self.capacity = capacity

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)

    def enqueue(self, vehicle_number, is_electric):
        if self.size() >= self.capacity:
            print(f"❌ Parkiran penuh! {vehicle_number} tidak bisa masuk dan harus menunggu.")
            return

        if is_electric:
            self.queue.appendleft(vehicle_number)  # Kendaraan listrik diutamakan
            print(f"⚡ Kendaraan listrik {vehicle_number} mendapatkan parkiran, silahkan masuk!")
        else:
            self.queue.append(vehicle_number)
            print(f"🚗 Kendaraan {vehicle_number} masuk ke parkiran.")

    def dequeue(self):
        if self.is_empty():
            print("❌ Parkiran kosong, tidak ada kendaraan yang keluar.")
        else:
            kendaraan = self.queue.popleft()
            print(f"✅ Kendaraan {kendaraan} telah keluar dari parkiran.")

    def front(self):
        if self.is_empty():
            print("❌ Parkiran kosong, tidak ada kendaraan di barisan depan.")
        else:
            print(f"🚘 Kendaraan pertama dalam parkiran: {self.queue[0]}")

    def rear(self):
        if self.is_empty():
            print("❌ Parkiran kosong, tidak ada kendaraan di barisan belakang.")
        else:
            print(f"🚖 Kendaraan terakhir dalam parkiran: {self.queue[-1]}")

    def show_queue(self):
        if self.is_empty():
            print("🅿️ Parkiran kosong.")
        else:
            print("\n🚦 Kendaraan dalam parkiran:", " -> ".join(self.queue))


# Program utama
parkiran = Antrian_Layanan_Parkir()

while True:
    print("\n<------------------------->")
    print("📍 MENU PARKIR MALL:")
    print("1️⃣ Masuk Parkiran/Ambil Tiket")
    print("2️⃣ Keluar Parkiran")
    print("3️⃣ Lihat Kendaraan di Parkiran")
    print("4️⃣ Lihat Kendaraan Pertama")
    print("5️⃣ Lihat Kendaraan Terakhir")
    print("6️⃣ Keluar Tidak Jadi Parkir")
    print("<------------------------->")

    pilihan = input("👉 Pilih Menu Parkiran: ").strip()

    if pilihan == "1":
        vehicle_number = input("🔢 Masukkan Nomor Kendaraan: ").upper().strip()
        is_electric = input("⚡ Apakah kendaraan Anda listrik? (y/n): ").strip().lower() == "y"
        parkiran.enqueue(vehicle_number, is_electric)

    elif pilihan == "2":
        parkiran.dequeue()

    elif pilihan == "3":
        parkiran.show_queue()

    elif pilihan == "4":
        parkiran.front()

    elif pilihan == "5":
        parkiran.rear()

    elif pilihan == "6":
        print("👋 Terima kasih! Sampai Jumpa Kembali.")
        break

    else:
        print("⚠️ Pilihan tidak valid, silakan coba lagi.")
```

```python
from collections import deque

class CertificateQueue:
    def __init__(self, max_size=20):
        self.regular_queue = deque()  # Antrian reguler
        self.achiever_queue = deque()  # Antrian mahasiswa berprestasi
        self.max_size = max_size  # Kapasitas maksimal

    def is_empty(self):
        return not self.regular_queue and not self.achiever_queue

    def size(self):
        return len(self.regular_queue) + len(self.achiever_queue)

    def enqueue(self, student_id, is_achiever=False):
        if is_achiever:
            print(f"🏅 Mahasiswa {student_id} (berprestasi) langsung mendapatkan sertifikat tanpa antre!")
            return

        if self.size() >= self.max_size:
            print(f"❌ Antrian penuh! Mahasiswa {student_id} tidak bisa ditambahkan.")
            return

        self.regular_queue.append(student_id)
        print(f"📚 Mahasiswa {student_id} masuk ke antrian reguler.")

    def dequeue(self):
        if self.is_empty():
            print("❌ Tidak ada mahasiswa dalam antrian.")
            return

        if self.achiever_queue:
            student_out = self.achiever_queue.popleft()
            print(f"🎖️ Mahasiswa {student_out} dari jalur prestasi telah mengambil sertifikat.")
        else:
            student_out = self.regular_queue.popleft()
            print(f"✅ Mahasiswa {student_out} dari jalur reguler telah mengambil sertifikat.")

    def front(self):
        if self.is_empty():
            print("🆓 Antrian kosong.")
            return

        if self.achiever_queue:
            print(f"🏅 Mahasiswa pertama dalam jalur prestasi: {self.achiever_queue[0]}")
        else:
            print(f"📚 Mahasiswa pertama dalam jalur reguler: {self.regular_queue[0]}")

    def rear(self):
        if self.is_empty():
            print("🆓 Antrian kosong.")
            return

        if self.regular_queue:
            print(f"📚 Mahasiswa terakhir dalam jalur reguler: {self.regular_queue[-1]}")
        else:
            print(f"🏅 Mahasiswa terakhir dalam jalur prestasi: {self.achiever_queue[-1]}")

    def show_queue(self):
        if self.is_empty():
            print("🆓 Antrian kosong.")
            return

        print("\n📜 **Antrian saat ini:**")
        if self.achiever_queue:
            print("🏅 **Jalur Prestasi:**", " -> ".join(self.achiever_queue))
        if self.regular_queue:
            print("📚 **Jalur Reguler:**", " -> ".join(self.regular_queue))


# Inisialisasi sistem antrian
certificate_queue = CertificateQueue()

print("=== SISTEM ANTRIAN PENGAMBILAN SERTIFIKAT KAMPUS ===")
while True:
    print("\n<------------------------->")
    print("MENU Sistem Antrian:")
    print("1️⃣ Tambah Mahasiswa ke Antrian")
    print("2️⃣ Tampilkan Antrian Saat Ini")
    print("3️⃣ Tampilkan Mahasiswa Pertama dan Terakhir")
    print("4️⃣ Layani Mahasiswa Pertama")
    print("5️⃣ Informasi Antrian")
    print("6️⃣ Layani Semua Mahasiswa hingga Kosong")
    print("7️⃣ Keluar")
    print(f"\nAntrian kosong: {certificate_queue.is_empty()}")
    print("<------------------------->")

    pilihan = input("Pilih Menu Sistem Antrian: ")

    if pilihan == "1":
        student_id = input("Masukkan ID Mahasiswa: ").upper()
        is_achiever = input("Apakah Mahasiswa Berprestasi? (y/n): ").strip().lower() == "y"
        certificate_queue.enqueue(student_id, is_achiever)

    elif pilihan == "2":
        certificate_queue.show_queue()

    elif pilihan == "3":
        certificate_queue.front()
        certificate_queue.rear()

    elif pilihan == "4":
        certificate_queue.dequeue()

    elif pilihan == "5":
        print(f"Total Mahasiswa dalam Antrian: {certificate_queue.size()}")

    elif pilihan == "6":
        print("\n📢 Melayani semua mahasiswa hingga antrian kosong...")
        while not certificate_queue.is_empty():
            certificate_queue.dequeue()
        print("✅ Semua mahasiswa telah dilayani. Antrian kosong.")

    elif pilihan == "7":
        print("Terima kasih! Program selesai.")
        break

    else:
        print("❌ Pilihan tidak valid, silakan coba lagi.")
```
